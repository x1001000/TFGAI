# 問題二
## 函數的傳入(傳值或傳址)及傳出(return)的觀念

```
X = [1001000, 3.14, 5566]
Y = sorted(X)
```

```
X
```

```
Y
```

1. sorted()函數之所以命名為sorted，就是為了讓程式有好的可讀性，看字面就能猜到：Y是排序過了的X
2. sorted()函數的傳入被實作成傳值，因此sorted(X)時，X不會被更動
3. sorted()函數的return值是一個list，指派給/傳址給/命名為Y

```
Y = X.sort()
```

```
X
```

```
Y
```

```
type(Y)
```

1. sort()是list類別的一個方法，方法(method)的本質就是函數(function)，一樣有傳入和傳出，類別若先天上常需要某種處理(列如串列常需要排序、字串常需要大小寫轉換)，類別的設計者就會為類別實作所需的方法，便於呼叫使用。
2. sort()方法的傳入被實作成傳址，因此X.sort()時，X會被更動排序
3. sort()方法的傳出空值(沒有實作return的部分)，把空值存入Y，因此Y是個空值，因此type(Y)是NoneType類別
4. 空值也是個特別的常數(如同True和False)，見底下四個cell，應該能明白我說的
