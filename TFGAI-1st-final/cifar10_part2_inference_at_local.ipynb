# Inference

8. 將cifar10_params_model.h5放在同一個工作目錄

- 匯入之前訓練好的model

```python
from keras.models import load_model
model = load_model('cifar10_params_model.h5')
```

10. 用匯入的模型/神經網路做推論/預測

- 將30萬張圖的分類(0~9)結果記在prediction串列

- 先用註解掉的那行估算時間

```python
from time import time
start = time()
prediction = []
for i in range(1, 300001):
    im = plt.imread('test/%d.png'%i)
    prediction.append(model.predict_classes(im.reshape(1, 32, 32, 3))[0])
    #if i%1000 == 0: print(i)
print('It took', (time()-start)/60, 'minutes')
```

# Submission

13. 將sampleSubmission.csv放在同一個工作目錄
- 看看提交的範例格式

```python
submission = pd.read_csv('sampleSubmission.csv')
submission
```

15. 準備要查的字典

```python
thing = {0:"airplane", 1:"automobile", 2:"bird", 3:"cat", 4:"deer", 5:"dog", 6:"frog", 7:"horse", 8:"ship", 9:"truck"}
```

16. 用數字翻譯成英文的串列建DataFrame

```python
df = pd.DataFrame([ thing[prediction[i]] for i in range(300000) ])
df
```

17. submission的label那欄更新為df

```python
submission['label'] = df
submission
```

18. submission輸出成csv檔，不包含index欄

```python
submission.to_csv('submission.csv', index=False)
```
